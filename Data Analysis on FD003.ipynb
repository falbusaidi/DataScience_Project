```python
# import necessery package
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.read_csv('inputData/Train_FD003.csv')
df1.describe()
```

```python
df_100 = df1[df1['UNIT_ID']==1]
```

```python
# Analysis based on RUL

fig, axes = plt.subplots(nrows=21,ncols=1, figsize=(6,80))

for i in range(1,22):
    if i < 10:
        index='SENSOR_0'+str(i)
    else:
        index='SENSOR_'+str(i)
    axes[i-1].scatter(df_100['RUL'],df_100[index])
    axes[i-1].set_xlabel('RUL')
    axes[i-1].set_ylabel(index)
fig.tight_layout()
```
